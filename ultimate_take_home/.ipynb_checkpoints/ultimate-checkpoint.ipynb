```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
logins = pd.read_json('logins.json', convert_dates=True)
logins['counts'] = 1
logins.head()
```

```python
assert logins[logins['login_time'] == None].empty
assert logins[logins['login_time'] == np.nan].empty
assert logins[logins['login_time'] == ''].empty
```

```python
login_counts = logins.set_index('login_time').resample('15T').sum()
login_counts.head()
```

```python
login_feb = login_counts[login_counts.index.month == 2]
login_feb.plot()
plt.show()
```

```python
login_counts['weekday'] = login_counts.apply(lambda row: row.name.weekday(), axis=1)
```

```python
avg_weekly_counts = login_counts.groupby('weekday').mean()
avg_weekly_counts.plot()
plt.show()
```

```python
tot_daily_counts = login_counts[['counts']].resample('D').sum()
tot_daily_counts.plot()
plt.show()
```

```python
login_counts['time_of_day'] = login_counts.apply(lambda row: row.name.time(), axis=1)
avg_daily_counts = login_counts.groupby('time_of_day')[['counts']].mean()
avg_daily_counts.plot()
plt.show()
```

Ultimate's current problem seems to be that each city has different active hours, but the availability of drivers does not mirror demand due to the barrier between the two cities. The end goal of reimbursing toll costs is increase driver availability in each city. A good measure of driver availability is the wait time for the rider to be picked up by a driver; a lower wait time indicates better driver availability. Therefore, I would choose average wait time as a metric and seek to minimize it.
