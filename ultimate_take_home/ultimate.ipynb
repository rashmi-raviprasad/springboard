```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
```

### Part 1: Aggregating and visualizing logins

First, we load the data and add a column for counts.

```python
logins = pd.read_json('logins.json', convert_dates=True)
logins['counts'] = 1
logins.head()
```

We check for missing data by ensuring there are no empty or null entries.

```python
assert logins[logins['login_time'] == None].empty
assert logins[logins['login_time'] == np.nan].empty
assert logins[logins['login_time'] == ''].empty
```

Next, we resample the counts at 15 minute intervals.

```python
login_counts = logins.set_index('login_time').resample('15T').sum()
login_counts.head()
```

There are a large number of entries, so instead of plotting the whole dataframe we will just plot February's login counts to detect any patterns. There seems to be both a daily and weekly pattern.

```python
login_feb = login_counts[['counts']][login_counts.index.month == 2]
login_feb.plot(legend=False)
plt.xlabel('Day of month')
plt.ylabel('Counts')
plt.title('Login counts by day in February')
plt.show()
```

Next, we can observe the average counts for each day of the week. We will need to extract the day of the week from each timestamp. There is an increase in login counts during the weekends and a steep drop on Mondays.

```python
login_counts['weekday'] = login_counts.apply(lambda row: row.name.weekday(), axis=1)
```

```python
avg_weekly_counts = login_counts.groupby('weekday').mean()
avg_weekly_counts.plot(legend=False)
plt.xticks(np.arange(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.xlabel('Day of week')
plt.ylabel('Counts')
plt.title('Average login counts by weekday')
plt.show()
```

Then, we can plot the daily total counts to observe long term patterns. It seems that the login counts have been increasing over the last few months.

```python
tot_daily_counts = login_counts[['counts']].resample('D').sum()
tot_daily_counts.plot(legend=False)
plt.xlabel('Day')
plt.ylabel('Counts')
plt.title('Total counts by day')
plt.show()
```

Finally, we can observe the daily cycles by plotting the average count for each 15 minute interval. There seem to be peaks in the middle of the day, in the evening, and in the middle of the night. The lowest number of logins typically occurs in the early morning.

```python
login_counts['time_of_day'] = login_counts.apply(lambda row: row.name.time(), axis=1)
avg_daily_counts = login_counts.groupby('time_of_day')[['counts']].mean()
avg_daily_counts.plot()
plt.xlabel('Time')
plt.ylabel('Counts')
plt.title('Average counts by time of day')
plt.show()
```

### Part 2: Experimental Design

1. Ultimate's current problem seems to be that each city has different active hours, but the availability of drivers does not mirror demand due to the barrier between the two cities. The end goal of reimbursing toll costs is to increase driver availability in each city. A good measure of driver availability is the wait time for the rider to be picked up by a driver; a lower wait time indicates better driver availability. Therefore, I would choose average wait time as a metric and seek to minimize it.

2. 
    (A) In order to test the effectiveness of the proposed change, I would reimburse the toll cost for a portion of drivers and label it as a promotion/reward of some kind. Then I would measure the average wait time for their rides and compare that with the average wait time for rides where drivers were not reimbursed.

    (B) I would conduct a one-tailed z-test to determine if the mean wait time for the test group is less than that of the population mean. The test would be one-tailed because we are looking for a decrease in mean wait time. I would use a z-test because we know the population variance and we are trying to compare a sample's metric to the population's value.

    (C) The null hypothesis is that the mean wait time of the test group is equal to or greater than that of the population; the alternative hypothesis is that the mean wait time is less than that of the population. I would set $\alpha$ = 0.05. If p < 0.05, we can reject the null hypothesis and conclude that the mean wait time for drivers with reimbursed toll costs is less than that of drivers without reimbursement.

    A caveat about this test is that it will have to be performed over the course of a week or two because of the fluctuations in demand throughout the week. It would also be best to make sure there are no holidays/special events during the testing period as that may affect demand.

For cleaning, we can encode the categorical data using a dictionary. That way we can still include it in our models.

```python
city_dict = {"King's Landing":0, 'Astapor':1, 'Winterfell':2}
phone_dict = {'iPhone':0, 'Android':1, None:2}
ultimate_df['city'] = ultimate_df['city'].map(city_dict)
ultimate_df['phone'] = ultimate_df['phone'].map(phone_dict)
```

Next, we need to get the difference between the date the customer signed up and their last trip. If their last trip was five months or more after they signed up, that means they are still active into their sixth month.

```python
ultimate_df['last_trip_date'] = pd.to_datetime(ultimate_df['last_trip_date'])
ultimate_df['signup_date'] = pd.to_datetime(ultimate_df['signup_date'])
```

I will then create a boolean column to indicate whether the customer is active into their sixth month. This column will serve as our labels.

```python
ultimate_df['month_diff'] = ((ultimate_df['last_trip_date'] - ultimate_df['signup_date'])/np.timedelta64(1, 'M')).astype(int)
ultimate_df['is_active'] = ultimate_df['month_diff'] >= 5
ultimate_df['is_active'] = ultimate_df['is_active'].astype(int)
ultimate_df.head()
```

Defining some helper functions for visualizing model metrics.

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix

def plot_roc_curve(model, X_test, y_test, name):
    probs = model.predict_proba(X_test)
    probs = probs[:, 1]
    no_skill = np.zeros(len(y_test))
    tpr, fpr, thresh = roc_curve(y_test, probs, pos_label=0)
    ns_tpr, ns_fpr, ns_thresh = roc_curve(y_test, no_skill, pos_label=0)
    plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill, AUC = %.2f' % auc(ns_fpr, ns_tpr))
    plt.plot(fpr, tpr, marker='.', markersize=0.1, label='%s, AUC = %.2f' % (name, auc(fpr, tpr)))
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.title('ROC Curve for %s' % name, fontsize=18)
    plt.legend()
    plt.show()

def plot_feature_importance(model, X):
    feature_ranks = np.flip(np.argsort(model.feature_importances_))
    feature_names = []
    for i in feature_ranks:
        feature_names.append(X.columns[i])
    sns.barplot(x=feature_names, y=np.flip(np.sort(model.feature_importances_)))
    plt.xticks(rotation=45, fontsize=14, ha='right')
    plt.yticks(fontsize=14)
    plt.xlabel('Feature name', fontsize=16)
    plt.ylabel('Feature importance', fontsize=16)
    plt.title('Features Ranked', fontsize=18)
    plt.show()

def plot_confusion_matrix(model, X_test, y_test):
    cm = confusion_matrix(y_test, model.predict(X_test))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
    plt.xticks(ticks=[0.5, 1.5], fontsize=14)
    plt.yticks(ticks=[0.5, 1.5], fontsize=14)
    plt.xlabel('Predicted labels', fontsize=16)
    plt.ylabel('True labels', fontsize=16)
    plt.title('Confusion Matrix', fontsize=18)
    plt.tight_layout()
    plt.show()
```

I dropped unnecessary columns before feeding the data into the model. I also completely removed the "avg_rating_of_driver" column because 16% of the data in that column was NaN. I did not want to remove 16% of all the records, so I just removed the column. If the company specifically wants to investigate this feature, I can perform this analysis using only records with values for this column.

I also dropped any other NaN entries in the dataframe. We are now left with 49,799 entries out of the original 50,000.

```python
from sklearn.model_selection import train_test_split

dropped_df = ultimate_df.drop(['avg_rating_of_driver', 'last_trip_date', 'signup_date', 'month_diff'], axis=1).dropna()
y_dropped = dropped_df['is_active']
X_dropped = dropped_df.drop('is_active', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_dropped, y_dropped, random_state=16, test_size=0.25)
```

#### 2. Modeling

Because this is a classification problem, I started with a logistic regression model. The accuracy scores were not bad but I received a warning message, so I decided to try other methods. If nothing else worked, I would come back to this model and address the warning; however, I wanted to try out other options first.

```python
from sklearn.linear_model import LogisticRegression

log_mod = LogisticRegression(random_state=16).fit(X_train, y_train)
print('Train accuracy:', log_mod.score(X_train, y_train), 'Test accuracy:', log_mod.score(X_test, y_test))
```

Next, a random forest classifier. Here, the training accuracy is very high but the testing accuracy is lower. This is clear evidence of overfitting. We will need to try something else.

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

rf_mod = RandomForestClassifier(random_state=16).fit(X_train, y_train)
print('Train accuracy:', rf_mod.score(X_train, y_train), 'Test accuracy:', rf_mod.score(X_test, y_test))
```

Finally, I tried a gradient boosting classifier. The train and test accuracy are both high, so I chose to proceed with this model.

```python
gb_mod = GradientBoostingClassifier(random_state=16).fit(X_train, y_train)
print('Train accuracy:', gb_mod.score(X_train, y_train), 'Test accuracy:', gb_mod.score(X_test, y_test))
```

The ROC curve for the gradient boosting model is below. The AUC is 0.82, which is quite high.

```python
plot_roc_curve(gb_mod, X_test, y_test, 'Gradient Boosting')
```

Below is the confusion matrix.

```python
plot_confusion_matrix(gb_mod, X_test, y_test)
```

Below is the feature importance ranking based on the gradient boosting model.

```python
plot_feature_importance(gb_mod, X_dropped)
```
